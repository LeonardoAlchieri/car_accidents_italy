```r
binomial  <- function(x,n,p)
    {
        factorial(n)/(factorial(n-x)*factorial(x)) * (p)^x * (1-p)^(n-x)
    }
```

```r
test <- seq(0,5,0.01)
results <- binomial(test, 5, 0.4)
```

```r
plot(test, results, type="l")
```

```r
result_sum_1 = 0
for(i in c(1:5))
    {
        result_sum_1 = binomial(i,5, 0.4) + result_sum_1
    }
result_sum_1
```

```r
result_sum_2 = 0
for(i in c(0:1))
    {
        result_sum_2 = binomial(i,4, 0.05) + result_sum_2
    }
result_sum_2
```

## Esercizio

Variabile casuale con $\mu = 30$cm, $\sigma^2 = 3.24$cm². Calcolo la probabilità che:
- $> 32$cm
- $= 29$
- $29 < x < 32$
- Ipotizzo $\mu = 31$, $P(X > 32)$
- Ipotizzo $\sigma^2 = 4$, $P(X > 32)$

**NOTA** Per ottenere la probabilità della Gaussiana $P(x < X)$, si utilizza in `R` la funzione `pnorm(X, MU, SIGMA)`.

```r
normal_dist <- function(x, mu, sigma)
    {
        1/(sigma*sqrt(2*pi)) * exp(-(x-mu)^2/(2*sigma^2))
    }
```

```r
MU = 30
SIGMA = sqrt(3.24)
x = seq(20, 40, 0.01)
```

```r
plot(x, normal_dist(x, MU, SIGMA), type="l")
```

```r
pnorm(32, mean=30, sd=sqrt(3.24), lower.tail = F)
```

```r
pnorm(29, mean=30, sd=sqrt(3.24), lower.tail = F) - pnorm(32, mean=30, sd=sqrt(3.24), lower.tail = F)
```

```r
pnorm(32, mean=31, sd=sqrt(3.24), lower.tail = F)
```

```r
pnorm(29, mean=30, sd=sqrt(3.24), lower.tail = F)
```

Abbiamo visto anche le distribuzioni del **chi-quadro**, **S** e **Student's t**. Si è poi mostrata la **Legge dei Grandi Numeri** e il **Teorema del Limite Centrale**.

Proviamo a definire una distribuzione binomiale. 

```r
N = 30
y1 <- dbinom(0:N, N, 1/4)
y2 <- dbinom(0:N, N, 1/2)
y3 <- dbinom(0:N, N, 3/4)
par(mfrow=c(1,3))
plot(y1, type="h")
plot(y2, type="h")
plot(y3, type="h")
```

Per calcolare la funzione di partizione della binomiale, si usa `pbinom`.

Per avere $P(X > 5)$ si fa

```r
1- pbinom(5,N, 1/4)
```

Con la funzione `qbinom` si calcola il *quantile*.

```r
qbinom(0.745, N, 1/2)
```

È stata poi mostrata la variabile casuale di **Poisson**, che si comporta in maniera analoga.
